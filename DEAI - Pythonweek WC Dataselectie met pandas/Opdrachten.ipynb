## Selecties op één tabel zonder functies

Geef een overzicht met daarin de producten en hun productiekosten waarvan de productiekosten lager dan 100 dollar en hoger dan 50 dollar ligt. (2 kolommen, 23 rijen)

```python
#1. Lees het dataFrame in
#2. We willen alles selecteren van deze tabel.
#3. We maken een boolean aan en kijken in de kolom 'PRODUCT_COST' 
#4. Waarin we productiekosten < 100 & > 50 willen projecteren op het scherm

product = pd.read_sql_query(select_all_product, conn)

product_kosten_hoger_lager = ((product['PRODUCTION_COST'] > 50) & (product['PRODUCTION_COST'] < 100))

product.loc[(product_kosten_hoger_lager), ['PRODUCT_NAME', 'PRODUCTION_COST']]




```

Geef een overzicht met daarin de producten en hun marge waarvan de marge lager dan 20 % of hoger dan 60 % ligt. (2 kolommen, 7 rijen) 

```python
#1. Lees de query in om een variabele te maken
#2. Maak een boolean waar marge < 20% of hoger dan 60%
#3. gebruik .loc[(boolean), ['kolommen']]

product = pd.read_sql(select_all_product, conn)

marge_hoger_lager = ((product['MARGIN'] < 0.2)) | ((product['MARGIN'] > 0.6))

product.loc[marge_hoger_lager, ['PRODUCT_NAME', 'MARGIN']]
```

Geef een overzicht met daarin de landen waar met francs wordt betaald. Sorteer de uitkomst op land.  (1 kolom, 3 rijen)

```python
#1. Lees de query om een variabele te maken
#2. Ik neem aan de we het tabel land gaan inlezen
#3. Maak een boolean met == 'francs'
#4. variabele.loq[bool, [kolom]]

land = pd.read_sql(select_all_country, conn)

land_betaald_met_francs = ((land['CURRENCY_NAME']) == 'francs')

land.loc[land_betaald_met_francs, ['COUNTRY']]

```

Geef een overzicht met daarin de verschillende introductiedatums waarop producten met meer dan 50% marge worden geïntroduceerd (1 kolom, 7 rijen) 

```python
# Lees de query om een variabele aan te maken
# Maak een boolean waarin je marge met meer dan 50% selecteert.
# Kolommen 'INTRODUCTION_DATE' en 'MARGIN'
# Sinds we verschillende waardes willen moeten we drop_duplicates gebruiken

product = pd.read_sql(select_all_product, conn)

marge_meer_dan_vijftig = ((product['MARGIN']) > 0.5)

product.loc[marge_meer_dan_vijftig, ['INTRODUCTION_DATE']].drop_duplicates('INTRODUCTION_DATE')
```

Geef een overzicht met daarin het eerste adres en de stad van verkoopafdelingen waarvan zowel het tweede adres als de regio bekend is (2 kolommen, 7 rijen)

```python
# Lees de query om een variabele aan te maken
# Welk tabel? sales_branch
# Welk kolommen heb je nodig?  ADRESS1, ADRESS2, REGION, CITY
# Maak een boolean waarin ADRESS2 != NaN en REGION NaN
# Dit doen we met notna()

salesbranch = pd.read_sql(select_all_sales_branch, conn)

alle_adressen_bekend = ((salesbranch['ADDRESS2'].notna()) & (salesbranch['REGION'].notna()))

salesbranch.loc[alle_adressen_bekend, ['ADDRESS1', 'CITY']]

```

Geef een overzicht met daarin de landen waar met dollars (dollars of new dollar) wordt betaald. Sorteer de uitkomst op land. (1 kolom, 4 rijen) 

```python
#Lees de query om een variabele aan te maken
# Welk tabel hebben we nodig? #country
# Welke kolommen hebben we nodig? #CURRENCY_NAME #COUNTRY
# Maak een boolean waar de landen met dollars of new dollar betalen.
# Sorteer de uitkomst op land. .sort_values

country = pd.read_sql(select_all_country, conn)

country_dollar_new_dollar = ((country['CURRENCY_NAME'] == 'dollars') | (country['CURRENCY_NAME'] == 'new dollar'))

country.loc[country_dollar_new_dollar, ['COUNTRY']].sort_values('COUNTRY')
```

Geef een overzicht met daarin beide adressen en de stad van vestigingen van klanten waarvan de postcode begint met een ‘D’ (van duitsland). Filter op vestigingen die een tweede adres hebben. (3 kolommen, 2 rijen) 

```python
# Lees de query.
# Welk tabel? Retailer_Site? 
# Welke kolommen? ADDRESS1, ADDRESS2, CITY, POSTAL_ZONE
# Maak een boolean postcode begint met 'd' filter op vestiging die twee adressen hebben.
# Uitkomst, beide adressen en de stad.

vestiging = pd.read_sql(select_all_retailer_site, conn)

postcode_begint_met_d = ( (vestiging['ADDRESS1'].notna()) & (vestiging['ADDRESS2'].notna()) & (vestiging['POSTAL_ZONE'].str[0] == 'D') )

vestiging.loc[postcode_begint_met_d, ['ADDRESS1', 'ADDRESS2', 'CITY']]
```

Maak 3 variabelen:
- Een met de laagste
- Een met de hoogste
- Een met de gemiddelde (afgerond op 2 decimalen)

marge van producten (3 kolommen, 1 rij) 

```python
# Lees de query in
# import numpy as np?
# Welke tabel? product
# Welke kolommen? MARGIN
# Maak een boolean met de laagste
# Maak een boolean met de hoogste
# Maak een boolean met het gemiddelde.
# Zet die om in een data frame.

product = pd.read_sql_query(select_all_product, conn)

laagste_margin = ((product['MARGIN'].min()))

hoogste_margin = ((product['MARGIN'].max()))

gemiddelde_margin = round(product['MARGIN'].mean(), 2)

resultaat = pd.DataFrame({'Laagste_marge' : [laagste_margin], 'Hoogste_marge' : [hoogste_margin], 'Gemiddelde_marge' : [gemiddelde_margin]})

resultaat

```
