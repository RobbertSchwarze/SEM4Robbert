Als de naam van de prinses niet in de tabel staat wordt het volgende door de methode teruggegeven: naamloos.<br>

```python
print("Voer uw prinses in:")
prinses = str(input())

def zoekPrins(prinses):
    match prinses:
        case "Aurora":
            return "Philip"
        case "Ariël":
            return "Eric"
        case "Jasmine":
            return "Aladdin"
        case "Pocahontas":
            return "John"
        case "Belle":
            return "Beast"
        case "Tiana":
            return "Naveen"
        case "Mulan":
            return "Shang"
        case _:
            return "naamloos"

print(zoekPrins(prinses))
```

## While-loop

Schrijf de inhoud van de methode wachtwoord. Deze heeft geen parameters en laat de gebruiker twee wachtwoorden invoeren, waarna hij bekijkt of deze gelijk zijn. Zolang dat niet het geval is wordt er eerst Fout, probeer het opnieuw! uitgeprint en moet de gebruiker het opnieuw proberen.<br>Als de gebruiker het goed doet, komt er Succes! op het scherm te staan.

```python
# Methode wachtwoord, geen parameters
# Gebruiker voor twee wachtwoorden in
# Zijn twee wachtwoorden gelijk?
# Niet goed, 'Fout, probeer het opnieuw!'
# Wel goed, 'Succes!'

def wachtwoord():
    incorrect = True

    while(incorrect):
        wachtwoord = str(input())
        wachtwoord1 = str(input())

        if(wachtwoord == wachtwoord1):
            print("Succes!")
            incorrect = False
        else:
            print("Fout, probeer het opnieuw!")

wachtwoord()
```

Schrijf de inhoud van de methode tafels. De methode print alle tafels vanaf 1 (dus inclusief de tafel van 1) tot en met de tafel van het meegekregen getal (dus inclusief de tafel van getal).  Zet een witregel tussen alle tafels in. <br>

## Objectoriëntatie

Maak een minisysteem voor een kledingwinkel. Maak verschillende klassen voor verschillende soorten kledingstukken, waarbij sommige klassen geërfd worden van andere klassen.
- Maak een klasse Kledingstuk aan, waarin door de constructor een merk, een maat en een prijs geïnitialiseerd worden.
- Maak een klasse Broek en een klasse Shirt, beide erven van de klasse Kledingstuk. Voeg aan Broek een broek_type, en aan Shirt een shirt_type toe (d.m.v. constructorgebruik).
- Maak een klasse Spijkerbroek die erft van broek en een klasse Poloshirt die erft van shirt. Voeg aan Spijkerbroek een blauwtint toe, en aan Poloshirt een kraagsoort (d.m.v. constructorgebruik).
- Implementeer in elke klasse een beschrijvingmethode, waarmee alle informatie over het betreffende kledingstuk wordt afgedrukt.
- Maak van elke klasse een object aan en roep de beschrijvingmethode aan. Controleer nauwkeurig of alles naar behoren werkt.

```python
# 1. Maak een klasse kledingstuk aan. x
# In de constructor wordt een merk een maat en een prijs geïnitialiseerd worden. x
# 2.Maak een klasse Broek die erft van Kledingstuk, Broek krijgt een broek_type erbij x
# 3.Maak een klasse Spijkerbroek die erft van broek, voeg aan spijkerbroek een blauwtint toe. x
# 4.Maak een klasse Poloshirt die erft van shirt, voeg een kraagsoort toe.
# 5.Implementeer in elke klasse een beschrijvings methode, waarmee alle info over het betreffende kleding stuk wordt afgedrukt.
# 6. Maak van elk klasse een object, roep de beschrijvings methode aan.

class Kledingstuk:

    def __init__(self, merk, maat, prijs):
        self.merk = merk
        self.maat = maat
        self.prijs = prijs

    def toString(self):
        return f"Merk: {self.merk} Maat: {self.maat} Prijs: {self.prijs} "

class Broek(Kledingstuk):

    def __init__(self, merk, maat, prijs, broek_type):
        super().__init__(merk, maat, prijs)
        self.broek_type = broek_type
    
    def toString(self):
        return super().toString() + f"Broektype : {self.broek_type}"

class Shirt(Kledingstuk):

    def __init__(self, merk, maat, prijs, shirt_type):
        super().__init__(merk, maat, prijs)
        self.shirt_type = shirt_type
    
    def toString(self):
        return super().toString() + f"Shirt Type : {self.shirt_type}"

class Spijkerbroek(Broek):
    
    def __init__(self, merk, maat, prijs, broek_type):
        super().__init__(merk, maat, prijs, broek_type)
        self.kleur = "Blauw"
    
    def toString(self):
        return super().toString() + f"Kleur: {self.kleur}"

class Poloshirt(Shirt):
    def __init__(self, merk, maat, prijs, shirt_type, kraagsoort):
        super().__init__(merk, maat, prijs, shirt_type)
        self.kraagsoort = kraagsoort
    
    def toString(self):
        return super().toString() + f"Kraagsoort: {self.kraagsoort}"

# Kledingstuk.
kledingstuk = Kledingstuk("Carhart", 40, 89.99)
print(kledingstuk.toString())

#Broek
broek = Broek("Levis", 40, 95.00, "Oversized")
print(broek.toString())

#Spijkerbroek
spijkerbroek = Spijkerbroek("Tommy Hilfiger", 38, 120.00, "Skinny Jeans")
print(spijkerbroek.toString())

#Poloshirt
poloshirt = Poloshirt("Polo Ralph Lauren", "L", 200.00, "Polo", "Button-down kraag")
print(poloshirt.toString())
```
